```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
#  
df_values = [1, 2, 3, 5, 10, 20, 50]

# x  
x = np.linspace(0, 30, 1000)

#  
plt.figure(figsize=(12, 8))

#      PDF  
for df in df_values:
    y = chi2.pdf(x, df)
    plt.plot(x, y, label=f'df={df}')

plt.title('  (  )')
plt.xlabel('x')
plt.ylabel('   (PDF)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

#  
df_values = [1, 2, 5, 10, 20]

# x  
x = np.linspace(-5, 5, 1000)

#  
plt.figure(figsize=(12, 8))

#    t- PDF  
for df in df_values:
    y = t.pdf(x, df)
    plt.plot(x, y, label=f'df={df}')

#  
y_norm = t.pdf(x, np.inf)
plt.plot(x, y_norm, label='Normal distribution', linestyle='--', color='black')

plt.title('t- (  )')
plt.xlabel('x')
plt.ylabel('   (PDF)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

#  
dfn_values = [1, 2, 5, 10]  #  
dfd_values = [1, 2, 5, 10]  #  

# x  
x = np.linspace(0, 5, 1000)

#  
plt.figure(figsize=(12, 8))

#     F- PDF  
for dfn in dfn_values:
    for dfd in dfd_values:
        y = f.pdf(x, dfn, dfd)
        plt.plot(x, y, label=f'dfn={dfn}, dfd={dfd}')

plt.title('F- (  )')
plt.xlabel('x')
plt.ylabel('   (PDF)')
plt.legend()
plt.grid(True)
plt.show()
```
