## 이원배치분산분석(Two-Way)
- 두 개의 독립 변수가 종속 변수에 미치는 영향을 분석하는 통계 기법

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/funda_sampling.csv",index_col=0)
```

```python
import statsmodels.api as sm
import pandas as pd
from statsmodels.formula.api import ols

# 데이터 불러오기
data = df[["region","type_of_business","amount"]]

## Two-way ANOVA에 맞는 수식을 설정하는 코드
two_way_anova = ols('amount ~ C(region) + C(type_of_business) + C(region):C(type_of_business)'
           , data = data).fit()
 
## 결과 해석을 저장하는 코드
result = sm.stats.anova_lm(two_way_anova)
 
## 결과 요약
print(result)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
# Tukey's HSD 사후검정 수행
# `pairwise_tukeyhsd`는 단일 범주형 변수에 대해 수행
# 이를 위해 'amount'를 'region'별로 구분하여 분석

tukey = pairwise_tukeyhsd(endog=data['amount'], groups=data['region'], alpha=0.05)

# Tukey's HSD 결과 출력
print("\nTukey's HSD 사후검정 결과:")
print(tukey)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
# Tukey's HSD 사후검정 수행
# `pairwise_tukeyhsd`는 단일 범주형 변수에 대해 수행
# 이를 위해 'amount'를 'region'별로 구분하여 분석

tukey = pairwise_tukeyhsd(endog=data['amount'], groups=data['type_of_business'], alpha=0.05)

# Tukey's HSD 결과 출력
print("\nTukey's HSD 사후검정 결과:")
print(tukey)
```
